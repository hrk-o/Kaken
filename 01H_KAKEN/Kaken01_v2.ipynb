## 研究-研究者データを作る
科研費研究ID,研究者ID,名前,所属,分野,職位,立場  
汚いコードだから目を瞑ってください

```python
import re
def namelist(kid, content, position):
    r = []
    r.append(kid)
    content = content.split(',')
    if len(content) == 1:#?
        if len(re.findall(r'([0-9]{8})' , content[0])) > 0:
            r.append(re.findall(r'([0-9]{8})' , content[0])[0])
        elif len(re.findall(r'([0-9]{7})' , content[0])) > 0:
            r.append(re.findall(r'([0-9]{7})' , content[0])[0])
        else:
            r.append('')
        r.append(content[0].split('(')[0].replace(' ',''))
        #r.append(content[0])
        r.append('')
        r.append('')
        r.append('')
    elif len(content) == 2:#学部がない人の場合
        if len(re.findall(r'([0-9]{8})' , content[1])) > 0:
            r.append(re.findall(r'([0-9]{8})' , content[1])[0])
        elif len(re.findall(r'([0-9]{7})' , content[1])) > 0:
            r.append(re.findall(r'([0-9]{7})' , content[1])[0])
        else:
            r.append('')
        #r.append(content[1][-9:-1])
        r.append(content[0][:-len(content[0].split(' ')[-1])].replace(' ',''))
        r.append(content[0].split(' ')[-1].replace(' ',''))
        r.append('')
        if len(re.findall(r'[^\x01-\x7E|A-Z]+', content[1])) > 0:
            r.append(re.findall(r'[^\x01-\x7E|A-Z]+', content[1])[0])
        else:
            r.append('')
        #r.append(content[1][:-10].replace(' ',''))
    elif len(content) == 3:
        if len(re.findall(r'([0-9]{8})' , content[2])) > 0:
            r.append(re.findall(r'([0-9]{8})' , content[2])[0])
        elif len(re.findall(r'([0-9]{7})' , content[2])) > 0:
            r.append(re.findall(r'([0-9]{7})' , content[2])[0])
        else:
            r.append('')
        #r.append(content[2][-9:-1])
        r.append(content[0][:-len(content[0].split(' ')[-1])].replace(' ',''))
        r.append(content[0].split(' ')[-1].replace(' ',''))
        r.append(content[1].replace(' ',''))
        if len(re.findall(r'[^\x01-\x7E|A-Z]+', content[2])) > 0:
            r.append(re.findall(r'[^\x01-\x7E|A-Z]+', content[2])[0])
        else:
            r.append('')
    else:
        r.append('')
        r.append('')
        r.append('')
        r.append('')
        r.append('')
        #r.append(content[2][:-10].replace(" ",""))
    r.append(position)
    return r
```

```python
rstudy_name = ["科研費研究ID","研究者ID","名前","所属","分野","職位","立場"]
rstudy = []
```

```python
import re
for index, df in kaken_df.iterrows():
    for ind, position in enumerate(['代表研究者','研究分担者','連携研究者','研究協力者','特別研究員','外国人特別研究員','受入研究者']):
        if type(df[ind+4]) != float:
            contents = df[ind+4].split('\n')
            for content in contents:
                rstudy.append(namelist(df[2], content, position))
```

```python
rstudy_pd = pd.DataFrame(rstudy, columns=rstudy_name)
rstudy_pd.to_csv('1996-2019_C_kakenstudy.csv', index=False, encoding='utf_8_sig')
```
